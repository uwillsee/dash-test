```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('Artworks.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data=data[['Title', 'Artist', 'Nationality', 'Gender', 'DateAcquired','Classification', 'Department','Medium']]
```

```python
data.info()
```

```python
# drop all rows with any empty space
data = data.dropna()
```

```python
data.info()
```

```python
# select rows with only one artist
data = data[~data.Artist.str.contains(", ")]
```

```python
data
```

```python
len(data.Artist.unique())
```

```python
data = data.drop_duplicates()
```

```python
data.info()
```

```python
def remove_punct(test_str, punc):
    # replace any punctuation received by an empty space
    # punc: list of strings
    for ele in test_str:  
        if ele in punc:  
            test_str = test_str.replace(ele, "") 
    return test_str
```

```python
# remove parentheses from nationality and gender 
punc=['(',')']
data['Nationality'] = data['Nationality'].map(lambda word: remove_punct(word,punc))
data['Gender'] = data['Gender'].map(lambda word: remove_punct(word,punc))
```

```python
# drop empty spaces = previously '()'
data.replace('',np.nan,inplace=True)
data = data.dropna()
```

```python
data.info()
```

```python
data.reset_index(inplace=True)
```

```python
# only take year from acquired date
data['DateAcquired'] = data['DateAcquired'].astype('datetime64[ns]').dt.year
```

```python
data['Gender'].unique()
```

```python
# treat gender categories
data['Gender'] = data['Gender'].replace({'male':'Male', 'female':'Female', 'Non-binary':'Non-Binary'})
```

```python
# select only female and male 
data = data[data['Gender'] != 'Non-Binary']
```

```python
# check duplicates artworks
data.loc[data.duplicated(subset=['Title', 'Artist']),:]
```

```python
# keep the first one only 
data.drop_duplicates(subset=['Title', 'Artist'], keep='first',inplace=True)
```

```python
# transform nationalities to countries
demonyms = pd.read_csv('demonyms.csv', header = None, index_col = 0)
other_demonyms = pd.DataFrame(['Other', 'Korea', 'Canada', 'Benin', 'United States'], columns = [1], index = ['Nationality unknown', 'Korean', 'Canadian Inuit', 'Beninese', 'Native American'])
demonyms = demonyms.append(other_demonyms)
data['Country'] = [str(demonyms.loc[nation, 1]) for nation in data['Nationality']]
```

```python
data['Country'].unique()
```

```python
# treat mexico
data.replace('Mexican    Mexico\nMexican    Mexico\nName: 1, dtype: object', 'Mexico', inplace = True)
```
